```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

def generate_synthetic_data(n_records):
    # Generate synthetic data
    # Generate Patient IDs
    patient_ids = np.arange(1, n_records + 1)

    # Generate random Diagnoses (0 = incorrect, 1 = correct)
    diagnoses = np.random.choice([0, 1], size=n_records, p=[0.1, 0.9]).astype(float)

    # Generate random Ages (between 0 and 100), allowing for NaN
    ages = np.random.randint(0, 100, size=n_records).astype(float)

    # Generate Hospital names (from a set of 5 hospitals)
    hospitals = np.random.choice(['Hospital_A', 'Hospital_B', 'Hospital_C', 'Hospital_D', 'Hospital_E'], size=n_records)

    # Generate Dates of Visit (within the last year)
    dates_of_visit = pd.to_datetime('2023-08-20') - pd.to_timedelta(np.random.randint(0, 365, size=n_records), unit='D')

    # Generate random Medical Codes (80% valid, 20% invalid)
    medical_codes = ['ICD-10-' + str(np.random.randint(1, 1000)) for _ in range(n_records)]
    invalid_codes = random.sample(range(n_records), int(0.2 * n_records))
    for idx in invalid_codes:
        medical_codes[idx] = 'INVALID'

    # Introduce some missing values in the dataset
    completeness = np.random.choice([0, 1], size=n_records, p=[0.1, 0.9])
    for i in range(n_records):
        if completeness[i] == 0:
            field_to_nullify = np.random.choice(['diagnosis', 'age', 'hospital', 'medical_code'])
            if field_to_nullify == 'diagnosis':
                diagnoses[i] = np.nan
            elif field_to_nullify == 'age':
                ages[i] = np.nan
            elif field_to_nullify == 'hospital':
                hospitals[i] = np.nan
            elif field_to_nullify == 'medical_code':
                medical_codes[i] = np.nan

    # Create DataFrame
    data = pd.DataFrame({
        'Patient_ID': patient_ids,
        'Diagnosis': diagnoses,
        'Age': ages,
        'Hospital': hospitals,
        'Date_of_Visit': dates_of_visit,
        'Medical_Code': medical_codes
    })

    return data

# Generate synthetic dataset
data = generate_synthetic_data(1000)

# Handle missing values
data_cleaned = data.dropna()

# Calculate Quality Metrics
# Accuracy: Percentage of correct diagnoses
accuracy = data_cleaned['Diagnosis'].mean() * 100

# Completeness: Ratio of fully documented patient records
completeness_ratio = data_cleaned.shape[0] / data.shape[0] * 100

# Consistency: Check for conflicting medical codes across hospitals
consistency_issues = data_cleaned.groupby('Medical_Code')['Hospital'].nunique().gt(1).sum()
consistency = (1 - consistency_issues / len(data_cleaned['Medical_Code'].unique())) * 100

# Timeliness: Average age of data entries (in days)
timeliness = (pd.to_datetime('2024-08-20') - data_cleaned['Date_of_Visit']).dt.days.mean()

# Validity: Percentage of valid medical codes
validity = (data_cleaned['Medical_Code'] != 'INVALID').mean() * 100

# Display results
metrics = {
    'Accuracy': accuracy,
    'Completeness': completeness_ratio,
    'Consistency': consistency,
    'Timeliness': timeliness,
    'Validity': validity
}

# Assign weights to each quality dimension
weights = {
    'Accuracy': 0.3,
    'Completeness': 0.25,
    'Consistency': 0.15,
    'Timeliness': 0.1,
    'Validity': 0.2
}

# Calculate composite score
composite_score = sum(metrics[dim] * weights[dim] for dim in metrics.keys())

# Print metrics and composite score
print("Quality Metrics:")
for metric, value in metrics.items():
    print(f"{metric}: {value:.2f}")

print(f"Composite Quality Score: {composite_score:.2f}")

# Simulate data quality over 12 months
time_periods = 12
accuracy_over_time = np.random.normal(loc=accuracy, scale=5, size=time_periods)
completeness_over_time = np.random.normal(loc=completeness_ratio, scale=3, size=time_periods)
consistency_over_time = np.random.normal(loc=consistency, scale=2, size=time_periods)
timeliness_over_time = np.random.normal(loc=timeliness, scale=10, size=time_periods)
validity_over_time = np.random.normal(loc=validity, scale=4, size=time_periods)

# Plot control charts
plt.figure(figsize=(14, 10))

plt.subplot(3, 2, 1)
plt.plot(accuracy_over_time, marker='o')
plt.title('Accuracy Over Time')
plt.xlabel('Month')
plt.ylabel('Accuracy (%)')

plt.subplot(3, 2, 2)
plt.plot(completeness_over_time, marker='o')
plt.title('Completeness Over Time')
plt.xlabel('Month')
plt.ylabel('Completeness (%)')

plt.subplot(3, 2, 3)
plt.plot(consistency_over_time, marker='o')
plt.title('Consistency Over Time')
plt.xlabel('Month')
plt.ylabel('Consistency (%)')

plt.subplot(3, 2, 4)
plt.plot(timeliness_over_time, marker='o')
plt.title('Timeliness Over Time')
plt.xlabel('Month')
plt.ylabel('Timeliness (Days)')

plt.subplot(3, 2, 5)
plt.plot(validity_over_time, marker='o')
plt.title('Validity Over Time')
plt.xlabel('Month')
plt.ylabel('Validity (%)')

plt.tight_layout()
plt.show()
```

```python
# Adjust the calculate_composite_score function
def calculate_composite_score(accuracy, completeness, consistency, timeliness, validity, weights):
    return (accuracy * weights['Accuracy'] +
            completeness * weights['Completeness'] +
            consistency * weights['Consistency'] +
            timeliness * weights['Timeliness'] +
            validity * weights['Validity'])

# Sensitivity analysis function
def sensitivity_analysis(metrics, weights, perturbation_percent=0.05):
    original_composite_score = calculate_composite_score(
        accuracy=metrics['Accuracy'],
        completeness=metrics['Completeness'],
        consistency=metrics['Consistency'],
        timeliness=metrics['Timeliness'],
        validity=metrics['Validity'],
        weights=weights
    )
    
    robustness_results = {}

    for metric in metrics:
        perturbed_metrics = metrics.copy()
        perturbed_value = metrics[metric] * (1 + perturbation_percent)
        perturbed_metrics[metric] = perturbed_value
        perturbed_composite_score = calculate_composite_score(
            accuracy=perturbed_metrics['Accuracy'],
            completeness=perturbed_metrics['Completeness'],
            consistency=perturbed_metrics['Consistency'],
            timeliness=perturbed_metrics['Timeliness'],
            validity=perturbed_metrics['Validity'],
            weights=weights
        )
        sensitivity = abs(perturbed_composite_score - original_composite_score) / original_composite_score * 100
        robustness_results[metric] = sensitivity

    return robustness_results

# Calculate robustness
robustness = sensitivity_analysis(metrics, weights)

# Print robustness results
print("\nRobustness Analysis (Sensitivity to 5% Change):")
for metric, sensitivity in robustness.items():
    print(f"Sensitivity of {metric}: {sensitivity:.2f}%")
```
